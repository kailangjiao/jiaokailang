```python
from urllib.request import urlopen as lp 
req = lp('http://placekitten.com/g/200/300')  #urlopen将字符串地址转换为request对象
#然后在默认传给urlopen。所以可写为如下单元格中的形式    #它的参数url是字符串或request对象
cat_dog = req.read()


with open('cat_dog_200.jpg','wb') as f: #图片是由二进制数据组成的，所以用二进制把所收到的数据写
#入jpg格式
    f.write(cat_dog)
    f.close()
```
