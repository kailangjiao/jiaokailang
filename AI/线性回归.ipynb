```python
import keras
import numpy as np
from keras.models import Sequential
import matplotlib.pyplot as plt
#按顺序构成的模型
from keras.layers import Dense
#成功，即把这几个包导入
```

```python
#使用numpy生成100个随机点
x_data = np.random.rand(100)    #随机函数生成100个随机点
#加噪音,也是随机点
noise = np.random.normal(0,0.01,x_data.shape)#噪音形状与x_data一样
y_data = x_data*0.1+0.2 + noise #noise会使y_data有一个小的扰动

#显示随机点
plt.scatter(x_data,y_data) #生成散点图
plt.show()
```

```python
#建顺序模型网络
model = Sequential()
#在模型中添加全链接层
model.add(Dense(units = 1,input_dim = 1))
model.compile(optimizer = 'sgd',loss = 'mse')
#sgd:随机梯度下降法.
#mse:均方误差。
```

```python
#迭代方式训练模型
#训练3001个批次
for step in range(3001):
    #每次训练一个批次
    cost = model.train_on_batch(x_data,y_data)
    #每500个batch打印一个cost
    if step % 500 == 0 :
        print('cost',cost)
        
#打印权值和偏执值
w,b = model.layers[0].get_weights()
print('w:',w,'b:',b)
#此处只有一层

#x_data输入网络中，得到预测值y_pred
y_pred = model.predict(x_data)

#显示随机点
plt.scatter(x_data,y_data)
#显示预测结果
plt.plot(x_data,y_pred,'r-',lw = 3)  #lw：线宽
#预测w近似为0.1，很逼近。因为有噪音
plt.show()
```
